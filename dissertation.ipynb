```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#Data import
collisions=pd.read_csv('https://raw.githubusercontent.com/LIUYH0123/dissertation/main/dft-road-casualty-statistics-collision-2022.csv', low_memory=False)
vehicle=pd.read_csv('https://raw.githubusercontent.com/LIUYH0123/dissertation/main/dft-road-casualty-statistics-vehicle-2022.csv', low_memory=False)
collisions.head()
```

```python
vehicle.head()
```

```python
merged_data = pd.merge(collisions, vehicle, on='accident_index', how='outer')
print(merged_data.head())
```

```python
merged_data.head()
```

```python
# Data Cleaning
# Remove attribute columns that are not used in the study
collision = merged_data[['accident_severity', 'age_band_of_driver', 'sex_of_driver', 'road_surface_conditions', 'road_type', 'junction_control', 'light_conditions', 'day_of_week', 'weather_conditions', 'did_police_officer_attend_scene_of_accident']]
# Remove missing values
for col in collision.columns:
    collision = (collision[collision[col]!=-1])
#  Remove unknown values
for col in collision.columns:
    collision = (collision[collision[col]!=9])
collision = collision[collision['light_conditions'] != 7]
collision = collision[collision['sex_of_driver'] != 3]
# Combining values with the same meaning
collision['did_police_officer_attend_scene_of_accident'].replace(3, 2, inplace=True)
# Classify values
def reclassify_age_band(age_band):
    if age_band in [1, 2, 3, 4]:
        return 1
    elif age_band in [5, 6]:
        return 2
    elif age_band in [7, 8]:
        return 3
    elif age_band in [9, 10, 11]:
        return 4
collision['age_band_of_driver'] = collision['age_band_of_driver'].map(reclassify_age_band)
# Remove missing values
collision.dropna(inplace=True)
collision.head()
print(collision.head())
```

```python
collision.head()
```

```python
# Mapping accident severity codes to labels
severity_labels = {1: '1 - Fatal', 2: '2 - Serious', 3: '3 - Slight'}

# Grouping data by 'age_band_of_driver' and 'accident_severity' and calculating counts
severity_day_counts = collision.groupby(['age_band_of_driver', 'accident_severity']).size().unstack(fill_value=0)

# Plotting the stacked bar chart
ax = severity_day_counts.plot(kind='bar', stacked=True)

# Updating legend labels
handles, labels = ax.get_legend_handles_labels()
updated_labels = [severity_labels[int(label)] for label in labels]
ax.legend(handles, updated_labels, title='Accident Severity')

# Customizing the x-axis labels
ax.set_xticklabels(['below 20', '21 to 35', '36 to 55', 'Over 55'])

plt.title('Accident Severity by Age')
plt.xlabel('Age')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.show()
```

```python
from sklearn.model_selection import train_test_split
collision_ml = collision.drop('accident_severity',axis=1)
X_train, X_test, y_train, y_test = train_test_split(collision_ml.values,
collision['accident_severity'].values,test_size=0.30, random_state=99)
```

```python
# Calculate the number of different categories
severity_counts = collision['accident_severity'].value_counts()
# Drawing pie charts
plt.pie(severity_counts, labels=['Slight', 'Serious', 'Fatal'], autopct='%1.1f%%')
plt.title('Accident Severity Distribution')
plt.show()
```

```python
!pip install imbalanced-learn
from imblearn.over_sampling import SMOTE
from collections import Counter

# Creating SMOTE Objects
smote = SMOTE(random_state=42)
# Oversampling the training set
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Distribution of categories of statistically processed data
severity_counts_resampled = Counter(y_train_resampled)
plt.pie(severity_counts_resampled.values(), labels=['Slight', 'Serious', 'Fatal'], autopct='%1.1f%%')
plt.title('Accident Severity Distribution after SMOTE')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, y_train)
Y_pred = decision_tree.predict(X_test)
acc_decision_tree1 = round(decision_tree.score(X_test, y_test) * 100, 2)
sk_report = classification_report(
    digits=6,
    y_true=y_test,
    y_pred=Y_pred)
print("Accuracy", acc_decision_tree1)
print(sk_report)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=200)
random_forest.fit(X_train,y_train)
Y_pred = random_forest.predict(X_test)
random_forest.score(X_test, y_test)
acc_random_forest1 = round(random_forest.score(X_test, y_test) * 100, 2)

sk_report = classification_report(
    digits=6,
    y_true=y_test,
    y_pred=Y_pred)
print("Accuracy" , acc_random_forest1)
print(sk_report)
```

```python
from xgboost import XGBClassifier
# Adjustment of category labels
y_train_adjusted = y_train - 1
y_test_adjusted = y_test - 1
# Creating and fitting an XGBoost classifier
xgb_classifier = XGBClassifier(n_estimators=200)
xgb_classifier.fit(X_train, y_train_adjusted)
y_pred_adjusted = xgb_classifier.predict(X_test)
accuracy = round(xgb_classifier.score(X_test, y_test_adjusted) * 100, 2)
sk_report = classification_report(
    digits=6,
    y_true=y_test_adjusted,
    y_pred=y_pred_adjusted
)
print("Accuracy" , acc_random_forest1)
print(sk_report)
```
